##Definición


Se llaman números de Fibonacci a aquellos que son términos de la sucesión de números naturales donde cada número se calcula sumando los dos anteriores a él.

(Definición sacada de: https://www.ecured.cu/Números_de_Fibonacci)


Esta sucesión se puede calcular de forma recursiva...

*Fibonacci(n)=Fibonacci(n-1)+Fibonacci(n-2)*

con los siguientes valores iniciales:

Fibonacci(0)=Fibonacci(1)=1

##Implementación y *analisis*

```
#se usa 'time' para calcular y guardar los datos del tiempo para las implementaciones
import time
import matplotlib.pyplot as plt
```

### Implementación recursiva

```
def fibonacci_recursivo(n):
  if n==0 or n==1:
    return 1
  else:
    return fibonacci_recursivo(n-1)+fibonacci_recursivo(n-2)
```

### Implementación secuencial o iterada

```
def fibonacci_secuencial(n):
  actual=0
  menos1=1
  menos2=1
  if n==0 or n==1:
    actual=1
  else:
    for i in range(2,n+1):
      actual=menos1+menos2
      menos2=menos1
      menos1=actual

  return actual

start=time.time()
s=fibonacci_secuencial(500)
end=time.time()
print(s)
print(end-start)
```

### Analisis de la implementación recursiva

Para analizar la función recursiva, se calculo el tiempo(en milisegundos) en ejecutarse la función en un rango de (0,50), y se comparo con una grafica lineal.

```
x=[]
y1=[]
n=50
for i in range(n):
  #calculando el tiempo de ejecución de la función
  start = time.time()
  fibonacci_recursivo(i)
  end = time.time()
  
  #agregando los valores para la grafica
  x.append(i)
  #end y start se encuentran en segundos, se hace la correspondiente conversión
  y1.append((end-start)*1000)

#graficando la grafica lineal
plt.plot(x,x,label='linear')
#graficando el tiempo de ejecucion de fibonacci
plt.plot(x,y1,label='Fibonacci recursivo')
plt.legend()
#limites para evitar que la función lineal se vea extraña
plt.xlim(0,n)
plt.ylim(0,n)
plt.xlabel("numero de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación recuriva de Fibonacci")
plt.show()
```

Como se puede ver, para valores pequeños, la implementación podria considerarse eficiente, pero la función cerca al n=15, comienza a comportarse como una función exponencial, por lo que se vuelve totalmente ineficiente.

### Analisis de la implementacion iterada

Al igual que para la implementación recursiva, para esta implementación se calculo su tiempo de ejecucion en milisegundos contra la cantidad de datos, ya que se sabe de antemano que esta implementación es, como minimo, algo mas eficiente que la recursiva, se probo con el rango de datos (0,500)

```
x=[]
y1=[]
n=500
for i in range(n):
  start = time.time()
  fibonacci_secuencial(i)
  end = time.time()
  
  x.append(i)
  #end y start se encuentran en segundos, se hace la correspondiente conversión
  y1.append((end-start)*1000)


#graficando la grafica lineal
plt.plot(x,x,label='linear')
#graficando el tiempo de ejecucion de fibonacci
plt.xlim(0,n)
plt.ylim(0,n)
plt.plot(x,y1,label='Fibonacci secuencial')
plt.legend()

plt.xlabel("numero de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación iterada de Fibonacci")
plt.show()
```

Ya que esta implementación es realmente eficiente, y como se ve, para un tiempo medido en milisegundos es basicamente impercebtible cualquier cambio.

Para poder visualizar la tabla de mejor manera, se opto por cambiar la medición del tiempo de milisegundo(x10^3) a nanosegundos(1x10^6)

```
x=[]
y1=[]
n=500
for i in range(n):
  start = time.time()
  fibonacci_secuencial(i)
  end = time.time()
  
  x.append(i)
  #end y start se encuentran en segundos, se hace la correspondiente conversión
  y1.append((end-start)*1000000)


#graficando la grafica lineal
plt.plot(x,x,label='linear')
#graficando el tiempo de ejecucion de fibonacci
plt.xlim(0,n)
plt.ylim(0,n)
plt.plot(x,y1,label='Fibonacci secuencial')
plt.legend()

plt.xlabel("numero de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación iterada de Fibonacci")
plt.show()
```

El ruido que se ve en la linea de "Fibonacci secuencial" se puede interpretar como variación en el tiempo interno del cpu, por lo que no es algo decisivo en relación a la eficiencia notable de esta implementación

###Comparación de ambas implementaciones

```
x=[]
y1=[]
y2=[]
#35 es un valor para el que la implementación recursiva no necesita mucho tiempo para ejecutarse
n=35
for i in range(n):
  x.append(i)

  # implementación secuencial
  start = time.time()
  fibonacci_secuencial(i)
  end = time.time()
  #end y start se encuentran en segundos, se hace la correspondiente conversión
  y1.append((end-start)*1000)

  # implementación recursiva
  start = time.time()
  fibonacci_recursivo(i)
  end = time.time()
  #end y start se encuentran en segundos, se hace la correspondiente conversión
  y2.append((end-start)*1000)

#graficando la grafica lineal
plt.plot(x,x,label='linear')

plt.xlim(0,n)
plt.ylim(0,n)
#graficando el tiempo de la función secuencial
plt.plot(x,y1,label='Fibonacci secuencial')
#graficando el tiempo de la función recursiva
plt.plot(x,y2,label='Fibonacci recursivo')
plt.legend()

plt.xlabel("numero de datos")
plt.ylabel("Tiempo en milisegundos")
plt.title("Implementación iterada de Fibonacci")
plt.show()
```

Como se nota a simple vista, la implementación iterada es, por mucho mas eficiente.

A simple vista, y comparando ambas implementaciones, la funcion recursiva sigue una curva exponencial, mientras que la secuancial una lineal
